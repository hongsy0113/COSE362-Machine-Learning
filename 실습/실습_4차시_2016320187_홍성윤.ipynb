## **가중합(Weighted Sum) 구하기**

입력과 가중치를 출력으로 변환하기 위해서는, 우선 각 입력에 대한 가중치를 더해서 가중합을 계산해야합니다.

가중합은 다음과 같은 수식으로 간단히 표현할 수 있습니다.

```
weighted sum = x_1 * w_1 + x_2 * w_2 + ... + x_n * w_n
```

위 수식을 구현한다고 하면 다음의 방식을 통해 계산할 수 있을 것입니다.

1. 가중합을 0으로 초기화합니다.
2. 첫 번째 입력에 해당 가중치를 곱하십시오. 이 결과를 가중합에 더해줍니다.
3. 다음 입력에 해당 가중치를 곱하십시오. 이 결과를 가중합에 더해줍니다.
4. 모든 입력에 대하여 위의 과정을 반복합니다.


## **Practice 2**

```weighted_sum``` 함수를 구현해봅시다.

1. 가중치 합의 값을 저장해 줄 ```weighted_sum```이라는 변수를 생성하고 0으로 초기화해줍시다.

    최종적으로 ```weighted_sum```은 해당 함수에서 반환해주어야 합니다.

2. 입력과 가중치의 쌍을 탐색한 후, 입력과 가중치를 곱해서 가중합에 더해줍시다.

````
# TODO: weighted_sum 메소드 구현하기
class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
  
  def weighted_sum(self, inputs):
    # TODO: 가중합을 저장하는 변수를 생성해주고 초기화해줍니다.
    weighted_sum = 0    
    for i in range(self.num_inputs):
      weighted_sum += inputs[i] * self.weights[i]
    # TODO: 가중합의 값을 함수의 결과로 반환해줍시다.
    return weighted_sum
````

* 입력에 대한 퍼셉트론의 가중합이 음수라면 -1을 반환합니다.

## **퍼셉트론 학습시키기**

이제 퍼셉트론은 주어진 입력 값을 예측할 수 있게 설계되었습니다. 퍼셉트론의 또 다른 강점은, 스스로 정답을 향해 학습을 수행한다는 것입니다.

```
training_set = {(18, 49): -1, (2, 17): 1, (24, 35): -1, (14, 26): 1, (17, 34): -1}
```

다음 코드는 우리가 퍼셉트론에 학습시킬 데이터를 생성시켜주는 코드입니다. 다음 코드를 실행시키고 해당 데이터의 특징을 살펴보도록 합시다.

````
import random

def generate_training_set(num_points):
	x_coordinates = [random.randint(0, 50) for i in range(num_points)]
	y_coordinates = [random.randint(0, 50) for i in range(num_points)]
	training_set = dict()
	for x, y in zip(x_coordinates, y_coordinates):
		if x <= 45-y:
			training_set[(x,y)] = 1
		elif x > 45-y:
			training_set[(x,y)] = -1
	return training_set
````

````
training_set = generate_training_set(30)
````

````
import seaborn
import matplotlib.pyplot as plt

x_plus = []
y_plus = []
x_minus = []
y_minus = []

for data in training_set:
	if training_set[data] == 1:
		x_plus.append(data[0])
		y_plus.append(data[1])
	elif training_set[data] == -1:
		x_minus.append(data[0])
		y_minus.append(data[1])
    
fig = plt.figure()
ax = plt.axes(xlim=(-25, 75), ylim=(-25, 75))

plt.scatter(x_plus, y_plus, marker = '+', c = 'green', s = 128, linewidth = 2)
plt.scatter(x_minus, y_minus, marker = '_', c = 'red', s = 128, linewidth = 2)

plt.title("Training Set")

plt.show()
````

    우선, 각 instance에 대한 퍼셉트론의 출력값을 예측해야 합니다. ```for``` 루프 내에서 ```predict```라는 변수를 만들고 ```.activation()```, ```.weighted_sum()``` 메서드를 활용해서 perceptron의 예측값을 해당 변수에 저장하도록 합시다.

2. 학습 데이터의 타겟값을 저장하는 변수 ```actual```을 생성한 후 값을 저장해주도록 합시다.

3. 변수 ```error```를 생성한 후에 실제 값과 예측 값의 차이를 계산해서 저장해주도록 합시다.

````
# TODO: training 메소드 구현해보기
class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
  
  def weighted_sum(self, inputs):
    weighted_sum = 0    
    for i in range(self.num_inputs):
      weighted_sum += inputs[i] * self.weights[i]
    return weighted_sum

  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    else :
      return -1

  def training(self, training_set):
    for inputs in training_set:
      prediction = self.activation(self.weighted_sum(inputs))
      actual = training_set[inputs]
      error = actual - prediction
````

## **가중치 수정하기**

앞서 우리는 데이터와 퍼셉트론의 예측값 사이의 오차를 구할 수 있었습니다.

이 오차를 지표로 삼아 우리는 퍼셉트론이 올바른 방향으로 나아갈 수 있도록 학습시킬 수 있습니다.

퍼셉트론의 예측, 즉 출력값은 입력과 가중치 곱의 총합인 가중합을 통해 결정되기 때문에, 퍼셉트론의 예측값을 변경하기 위해서는 출력값을 변경해줘야합니다. 이 중에서 우리가 임의로 변경할 수 있는 변수는 가중치입니다. 

퍼셉트론 학습의 목표는 가중치를 올바른 방향으로 수정해서, 정답을 출력하는 최적의 가중치를 찾는 것입니다.

그렇다면 우리는 어떻게 최적의 가중치를 찾을 수 있을까요? 

최적의 가중치를 찾는 방법 중, 가장 널리 사용되고 있는 것이 바로 경사하강법(Gradient Descent)입니다. 이번에 우리는 Step Activation Function을 사용하기 때문에, 경사하강법 중 Delta Rule을 통해 가중치를 학습시킬 것입니다.

tweaking_the_weights.png

Delta rule은 다음의 간단한 규칙을 통해 가중치를 수정하게 됩니다.

```
weight = weight + learning rate * (error * input)
```

## **Practice 5**

Delta Rule로 퍼셉트론을 학습시키기 위해 앞서 구현했던 training 메소드를 수정해봅시다.

1.우선 정답을 찾을 때까지 학습을 반복하기 위해 다음의 요소들을 추가해줍시다.

* ```foundLine = False``` (퍼셉트론이 양수 레이블과 음수 레이블을 구분하는 line을 찾았는지 여부를 나타내는 bool 변수)
* ```while not foundLine:``` (라인, 즉 정답을 찾을 때까지 퍼셉트론을 계속 훈련시키는 while 루프)
* ```total_error = 0``` (퍼셉트론이 각 라운드에서 만드는 총 오류를 계산)
* ```total_error += abs(error)``` (각 라운드에서 퍼셉트론이 만드는 총 오류를 업데이트하기 위해)

2. 알고리즘이 오류를 찾지 못하면 퍼셉트론이 모든 데이터 instance에 대해 정답을 올바르게 예측했다는 의미입니다.

      ```for``` 루프의 밖에서, ```total_error```가 ```0```이라면 ```foundLine``` 값을 ```True```로 바꿔줍시다.

3. 각각의 가중치에 대해 Delta Rule을 적용하여 각 가중치 ```self.weights[i]```를 업데이트합니다

    ```weight = weight + (error * inputs)```

주어진 ```small_training_set```을 이용해서 퍼셉트론을 학습시켜봅시다.

```.weights``` 내부 변수를 출력해서 학습된 가중치 수치 또한 직접 확인해보도록 합니다.

````
# TODO: training()메소드에 delta rule 추가
class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
  
  def weighted_sum(self, inputs):
    weighted_sum = 0    
    for i in range(self.num_inputs):
      weighted_sum += inputs[i] * self.weights[i]
    return weighted_sum

  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    else :
      return -1

  def training(self, training_set):
    foundLine = False
    while not foundLine:
      total_error = 0
      for inputs in training_set:
        prediction = self.activation(self.weighted_sum(inputs))
        actual = training_set[inputs]
        error = actual - prediction
        total_error += abs(error)

        for i in range(self.num_inputs):
          self.weights[i] += error*inputs[i]
      if total_error == 0:
        foundLine = True  
````

## **편향치(Bias)**

앞서 우리는 퍼셉트론이 가중치를 조정하여 올바른 출력을 생성하도록 학습 될 수 있음을 알 수 있었습니다.

그러나 퍼셉트론의 정확도를 높이기 위해 약간의 조정이 필요한 경우가 있습니다. 이를 위해 우리는 퍼셉트론에 편향치를 추가합니다.

편향이 추가된 가중합의 식은 다음과 같습니다.

```
weighted sum = x_1 * w_1 + x_2 * w_2 + ... + x_n * w_n​ + w_b
```

다음 두 가지 변경사항만 고려하면 앞서 구현한 퍼셉트론에 편향을 적용할 수 있을 것입니다.

* 입력 데이터의 마지막에 1을 추가합니다 (이제 2 대신 3 개의 입력이 있음).
* 가중치 목록에 편향 가중치를 추가합니다 (이제 2 개 대신 3 개의 가중치가 있음).

````
# TODO: __init__ 생성자를, 편향치가 적용될 수 있도록 적합하게 수정해봅시다.
# TODO: training()메소드에 delta rule 추가
class Perceptron:
  def __init__(self, num_inputs=3, weights=[1,1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
  
  def weighted_sum(self, inputs):
    weighted_sum = 0    
    for i in range(self.num_inputs):
      weighted_sum += inputs[i] * self.weights[i]
    return weighted_sum

  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    else :
      return -1

  def training(self, training_set):
    foundLine = False
    while not foundLine:
      total_error = 0
      for inputs in training_set:
        prediction = self.activation(self.weighted_sum(inputs))
        actual = training_set[inputs]
        error = actual - prediction
        total_error += abs(error)

        for i in range(self.num_inputs):
          self.weights[i] += error*inputs[i]
      if total_error == 0:
        foundLine = True  
````

## **퍼셉트론을 통한 선형 분류**

퍼셉트론은 예상대로 작동하지만 지금은 훈련 과정을 확인할 수 없습니다. 무슨 일이 일어나고 있는지 더 잘 이해하기 위해 퍼셉트론의 훈련 과정을 시각화 해봅니다.

가중치는 훈련 과정 전반에 걸쳐 변경되므로 해당 가중치를 의미있게 시각화 할 수만 있다면 기울기-절편 형식을 사용하여 선을 나타낼 수 있다는 것을 알고있을 것입니다. 퍼셉트론의 가중치는 퍼셉트론이 나타내는 선의 기울기와 절편을 찾는 데 사용할 수 있습니다.

* 기울기 = ```-self.weights[0] /self.weights[1]```
* 절편 = ```-self.weights[2] /self.weights[1]```

````
from matplotlib.colors import ListedColormap

# Perceptron의 decision boundary를 그려주는 함수입니다.
# 함수 출처: http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
def plot_decision_boundary(classifier, X, y, title):
  xmin, xmax = np.min(X[:, 0]) - 0.05, np.max(X[:, 0]) + 0.05
  ymin, ymax = np.min(X[:, 1]) - 0.05, np.max(X[:, 1]) + 0.05
  step = 0.01
  cm = plt.cm.coolwarm_r
  thr = 0.0
  xx, yy = np.meshgrid(np.arange(xmin - thr, xmax + thr, step), np.arange(ymin - thr, ymax + thr, step))
  if hasattr(classifier, 'decision_function'):
    Z = classifier.decision_function(np.hstack((xx.ravel()[:, np.newaxis], yy.ravel()[:, np.newaxis])))
  else:
    Z = classifier.predict_proba(np.hstack((xx.ravel()[:, np.newaxis], yy.ravel()[:, np.newaxis])))[:, 1]
  Z = Z.reshape(xx.shape)
  plt.contourf(xx, yy, Z, cmap=cm, alpha=0.8)
  plt.colorbar()
  plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['#FF0000', '#0000FF']), alpha=0.6)
  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)
  plt.xticks((0.0, 1.0))
  plt.yticks((0.0, 1.0))
  plt.title(title)
````
